```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd         
import tqdm
import time 
```

```python
page = 1
base_url = "https://dantri.com.vn/kinh-doanh/chung-khoan/trang-"
articles = []  

while True:     
    url = base_url + str(page) + ".htm"
    page += 1

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    if soup == None:
        break

    target = soup.find('div', class_='article list')

    for i in tqdm.tqdm(target.find_all('h3')):
        link_temp = "https://dantri.com.vn/" + i.find('a')['href']
        article_data = {
            'link': link_temp,
            'title': i.text.strip(),
            'time' : [],
            'content': []
        }
        articles.append(article_data)
        response = requests.get(link_temp)
        article_soup = BeautifulSoup(response.text, 'html.parser')
        article_data['time'] = article_soup.find('time', class_='author-time').text.strip()
        article_content = article_soup.find('div', class_='singular-content')

        if article_content:
            for p in article_content.find_all('p'):
                article_data['content'].append(p.text.strip())
        
            article_data['content'] = " ".join(article_data['content'])  
        articles.append(article_data)
        time.sleep(2)
# articles  
```

```python
data = pd.DataFrame(articles)
data['link']
```

```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import tqdm
import time 

url = "https://dantri.com.vn/kinh-doanh/chung-khoan/trang-"
page = 30
base_url = url
articles = []  

while True:
    try:
        url = base_url + str(page) + ".htm"

        page += 1
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        target = soup.find('div', class_='article list')

        # break code if there is no more article
        if target == None:
            break

        for i in tqdm.tqdm(target.find_all('h3')):
            article_data = {
                'link': "https://dantri.com.vn" + i.find('a')['href'],
                'title': i.text.strip(),
                'time' : [],
                'content': []
            }
            response = requests.get(article_data['link'])
            article_soup = BeautifulSoup(response.text, 'html.parser')
            article_data['time'] = article_soup.find('time', class_='author-time').text.strip().split(" ")[2]
            article_content = article_soup.find('div', class_='singular-content')   
            if article_content:
                for p in article_content.find_all('p'):
                    article_data['content'].append(p.text.strip())
                
                article_data['content'] = " ".join(article_data['content'])    

            articles.append(article_data)


    except Exception as e:
        print(e)
        continue
```
